```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10,10,100)
h  = 0.05

sqrt= lambda x: 1/(np.sqrt(1+(np.e**-x**2)))
error= (x*(np.e**-x**2))/(1+np.e**-x**2)**3/2

f = sqrt(x)

def Derivada(x,f,h):
    return (f(x+h)-f(x-h))/(2*h)

DerivadaC = Derivada(x,sqrt,h)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(2,2,1)
ax1 = fig.add_subplot(2,2,2)
ax2 = fig.add_subplot(2,2,3)

ax.scatter(x,f)
ax1.scatter(x,DerivadaC)
ax2.scatter(x,np.abs(error-DerivadaC))


print(Derivada(x,sqrt,h))
```
